```python
import matplotlib.pyplot as plt
import random
import numpy as np

from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False


#如果是 mac 系统,上面的这两个可能不管用,就用这个
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
```

# 单个曲线图

```python
x = range(30)
y_nanjing = [random.uniform(0,10) for i in x]

#dpi是清晰度度
plt.figure(figsize=(20,8), dpi=100)
#画图
plt.plot(x, y_nanjing)

#用这个字符串替代 x 轴上的数字
x_ticks_label=['12月{}日'.format(i) for i in x]
#设置 y 轴的显示范围
y_ticks = range(15)

#修改x,y 轴的坐标显示,xy 轴都是间隔 2,其中 x 周的数字用x_ticks_label中的字符串代替
plt.xticks(x[::2],x_ticks_label[::2], fontsize=15)
plt.yticks(y_ticks[::2], fontsize=15)

#alpha是透明度
plt.grid(True, linestyle='--', alpha=0.5)

#fontsize字体大小
plt.xlabel('time', fontsize=15)
plt.ylabel('temperature', fontsize=15)
plt.title('December temperature', fontsize=30)

#图像保存,一定要放在plt.show()前面,plt.show()之后,所有资源就都释放了
plt.savefig("./test.eps")

plt.show()
```

# 一个图里画多个曲线

```python
x = range(30)
y_nanjing = [random.uniform(0,10) for i in x]
y_wuhan = [random.uniform(5,15) for i in x]

plt.figure(figsize=(20,8), dpi=100)

plt.plot(x, y_nanjing, color='g', linestyle='-', label='nanjing')
plt.plot(x, y_wuhan, color='r', linestyle='--', label='wuhan')

#用这个字符串替代 x 轴上的数字
x_ticks_lable = ['12月{}日'.format(i) for i in x]
#设置 y 轴显示范围
y_ticks = range(16)

plt.xticks(x[::2], x_ticks_lable[::2], fontsize=15)
plt.yticks(y_ticks[::2], fontsize=15)

#alpha是透明度
plt.grid(True, linestyle='--', alpha=1)

plt.xlabel('time', fontsize=20)
plt.ylabel('temperature', fontsize=20)
plt.title('December temperature', fontsize=20)
#loc是位置参数,参数在这里:https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html
plt.legend(loc=0, fontsize=20)
#图像保存,一定要放在plt.show()前面,plt.show()之后,所有资源就都释放了
plt.savefig("./test.eps")

plt.show()
```

# 多个图片分开画

```python
x = range(30)
y_nanjing = [random.uniform(0,10) for i in x]
y_wuhan = [random.uniform(5,15) for i in x]
#注意花多个图的时候,nrows是行数,ncols 是列数
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=100)

#分别画两个图
axes[0].plot(x, y_nanjing, label='nanjing')
axes[1].plot(x, y_wuhan, label='wuhan', color='r', linestyle='--')

#用这个字符串替代 x 轴上的数字
x_ticks_lable = ['12月{}日'.format(i) for i in x]
#限制 y 轴的显示范围
y_ticks = range(16)

#x,y 轴都是间隔 2
axes[0].set_xticks(x[::2])
axes[0].set_yticks(y_ticks[::2])
#用x_ticks_lable字符串替代 x 轴上的数字
axes[0].set_xticklabels(x_ticks_label[::2])

axes[1].set_xticks(x[::2])
axes[1].set_yticks(y_ticks[::2])
axes[1].set_xticklabels(x_ticks_label[::2])

#alpha 是透明度
axes[0].grid(True, linestyle='--', alpha=1)
axes[1].grid(True, linestyle='--', alpha=1)

axes[0].set_xlabel('time')
axes[0].set_ylabel('temperature')
axes[0].set_title('december temperature', fontsize=20)

axes[1].set_xlabel('time')
axes[1].set_ylabel('temperature')
axes[1].set_title('december temperature', fontsize=20)

#图像保存,一定要放在plt.show()前面,plt.show()之后,所有资源就都释放了
plt.savefig('./test.eps')

plt.show()
```

# 散点图

```python
x = range(30)
y_nanjing = [random.uniform(0,10) for i in x]
plt.figure(figsize=(20,8), dpi=100)
plt.scatter(x, y_nanjing, color='r', label='price')
plt.grid()
plt.legend(loc=0, fontsize=20)
plt.show()
```

# 柱状图

```python
movie_name = ['神奇女侠', '超人', '钢铁侠']
x = range(len(movie_name))
y = [4, 6, 3]

plt.figure(figsize=(20, 8), dpi=100)
plt.bar(x, y, color=['r', 'g', 'b'], width=0.8)

plt.xticks(x, movie_name)
plt.grid(True, linestyle='--', alpha=0.8)

plt.title('电影票房')
plt.show()
```

# 直方图

```python


# 生成均匀分布的随机数
x2 = np.random.uniform(-1, 1, 100000000)

# 画图看分布状况
# 1）创建画布
plt.figure(figsize=(10, 10), dpi=100)

# 2）绘制直方图
plt.hist(x=x2, bins=1000)  # x代表要使用的数据，bins表示要划分区间数

# 3）显示图像
plt.show()
```
